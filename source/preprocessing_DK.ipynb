### Với các cột numeric, các giá trị được phân bố như thế nào?

Ở phần này, với các cột có kiểu dữ liệu `numeric`, ta sẽ tính:
+ Tỉ lệ phần trăm các giá trị thiếu.
+ Giá trị min, mean, max.
+ Giá trị tứ phân vị thứ nhất, trung vị và tứ phân vị thứ 3.

```python
def missing_ratio(df):
    return (df.isnull().sum() * 100 / len(df)).round(1)

def lower_quartile(df):
    return df.quantile(0.25).round(1)

def mean(df):
    return df.mean().round(1)

def median(df):
    return df.quantile(0.5).round(1)

def upper_quartile(df):
    return df.quantile(0.75).round(1)

num_col_info_df = train.select_dtypes(include=np.number)
num_col_info_df = num_col_info_df.agg([missing_ratio, "min", mean, lower_quartile, median, upper_quartile, "max"])
num_col_info_df
```

### Với các cột categorical, các giá trị được phân bố như thế nào?

Ở phần này, với các cột có kiểu dữ liệu `categorical`, ta sẽ tính:
+ Tỉ lệ phần trăm các giá trị thiếu.
+ Số lượng các giá trị khác nhau (ta sẽ không tính các giá trị thiếu).
+ Tỉ lệ phần trăm của mỗi giá trị trong cột.

```python
def num_values(df):
    return df.nunique()

def value_ratios(df):
    value_counts = df.value_counts(normalize=True)*100
    value_ratios = value_counts.round(1).to_dict()
    value_ratios = dict(sorted(value_ratios.items(), key=lambda item: item[1], reverse=True))
    return value_ratios

cat_col_info_df = train.select_dtypes(include='object').drop(columns=['Id','idhogar'])
cat_col_info_df = cat_col_info_df.agg([missing_ratio, num_values, value_ratios])
cat_col_info_df
```

Nhóm nhận thấy ý nghĩa của các cột `dependency`, `edjefe`, `edjefa` nên có dữ liệu số. Ta sẽ giải quyết vấn đề này theo phương pháp sau:
+ Đối với cột `dependency`, ta sẽ tính toán lại bằng cách lấy căn bậc 2 từ cột `SQBdependency`.
+ Đối với 2 biến trình độ giáo dục `edjefe` và `edjefa`, ta sẽ chuẩn hóa **yes = 1**, **no = 0** và với các giá trị còn lại ta sẽ chuyển sang dữ liệu số.

```python
mapping = {"yes": 1, "no": 0}

for df in [train, test]:
    df['dependency'] = np.sqrt(df['SQBdependency'])
    df['edjefa'] = df['edjefa'].replace(mapping).astype(np.float64)
    df['edjefe'] = df['edjefe'].replace(mapping).astype(np.float64)
```

### Cột meaneduc

`meaneduc` là số năm học trung bình của người lớn (hơn 18 tuổi).

```python
train.loc[pd.isnull(train['meaneduc'])].head()
```

Nhóm sẽ điền các giá trị thiếu bằng cách đi tính giá trị trung bình của số năm học `escolari` theo từng hộ gia đình.

```python
mean_escolari_by_idhogar = train.groupby('idhogar')['escolari'].mean()
train['meaneduc'] = train.apply(lambda row: mean_escolari_by_idhogar[row['idhogar']] if pd.isnull(row['meaneduc']) else row['meaneduc'], axis=1)

mean_escolari_by_idhogar = train.groupby('idhogar')['escolari'].mean()
train['meaneduc'] = train.apply(lambda row: mean_escolari_by_idhogar[row['idhogar']] if pd.isnull(row['meaneduc']) else row['meaneduc'], axis=1)
```

### Cột SQBmeaned

`SQBmeaned` là bình phương của `meaneduc` nên với những giá trị thiếu ở cột này nhóm sẽ đi tính lại giá trị bình phương và điền vào.

```python
train.loc[pd.isnull(train['SQBmeaned']), 'SQBmeaned'] = train.loc[pd.isnull(train['SQBmeaned']), 'meaneduc'] ** 2
test.loc[pd.isnull(test['SQBmeaned']), 'SQBmeaned'] = test.loc[pd.isnull(test['SQBmeaned']), 'meaneduc'] ** 2
```
